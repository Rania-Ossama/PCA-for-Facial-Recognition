```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

# Load ALL columns (pixels)
df = pd.read_csv("X_TRAIN.csv")

print("DataFrame shape:", df.shape)

# Convert to NumPy array
X_train = df.values

print("X_train shape:", X_train.shape)
```

```python
print("NaN values in X_train:", np.isnan(X_train).sum())

# Replace NaN with 0
X_train = np.nan_to_num(X_train)

```

```python
mean_face = np.mean(X_train, axis=0)
print("Mean face shape:", mean_face.shape)
```

```python
plt.imshow(mean_face.reshape(100, 100), cmap="gray")
plt.title("Mean Face (Average Face)")
plt.axis("off")
plt.show()
```

```python
X_centered = X_train - mean_face
print("Centered data shape:", X_centered.shape)

```

```python
L = np.dot(X_centered, X_centered.T)

print("Covariance matrix shape:", L.shape)
print("NaN values in covariance matrix:", np.isnan(L).sum())
```

```python
np.save("X_centered.npy", X_centered)
np.save("mean_face.npy", mean_face)
np.save("L.npy", L)
```
